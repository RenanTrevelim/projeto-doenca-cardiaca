## 🔎 Fatores Críticos Relacionados à Doença Cardíaca

Nesta etapa, foram analisadas variáveis consideradas clinicamente relevantes, com o objetivo de identificar padrões que diferenciam pacientes com e sem doença cardíaca. Para isso, foram utilizados boxplots, violin plots e curvas de densidade, permitindo uma análise visual mais completa das distribuições.


```python
def plot_triplo(dados, x, y, titulo_y=None):
    fig, ax = plt.subplots(1, 3, figsize=(16,4))

    palette = {
        'Presenca': '#2ecc71',
        'Ausencia': '#b0b0b0'
    }

    order = ['Presenca', 'Ausencia']

    # Boxplot
    sns.boxplot(
        data=dados, x=x, y=y,
        order=order, ax=ax[0],
        palette=palette, hue=x, legend=False
    )
    ax[0].set_title(f'Boxplot de {y}')
    ax[0].set_xlabel(x)
    ax[0].set_ylabel(y)

    # Violinplot
    sns.violinplot(
        data=dados, x=x, y=y,
        order=order, ax=ax[1],
        palette=palette, hue=x, legend=False
    )
    ax[1].set_title(f'ViolinPlot de {y}')
    ax[1].set_xlabel(x)
    ax[1].set_ylabel(y)

    # KDE plot
    sns.kdeplot(
        data=dados, x=y,
        hue=x, ax=ax[2],
        fill=True
    )
    ax[2].set_title(f'Densidade de {y}')
    ax[2].set_xlabel(y)

    plt.tight_layout()
    plt.show()
```

### ❤️ Tipo de Dor no Peito (`Tipo dor`)

A análise da variável relacionada ao tipo de dor no peito revela diferenças importantes entre os grupos:

- Pacientes com **doença cardíaca** tendem a apresentar valores mais altos nessa variável  
- A distribuição indica maior concentração em categorias associadas a dor mais significativa  
- Já os pacientes sem a doença apresentam maior dispersão em categorias intermediárias  

📌 **Insight:** O tipo de dor no peito se mostra um forte indicativo clínico, sendo um dos principais fatores associados à presença da doença.


```python
plot_triplo(dados, 'Doenca cardiaca', 'Tipo dor')
```

### 💓 Frequência Cardíaca Máxima (`Frequencia cardiaca max`)

A frequência cardíaca máxima atingida durante esforço apresenta um comportamento distinto:

- Pacientes com doença cardíaca tendem a apresentar **valores mais baixos**  
- Já os indivíduos sem a doença apresentam **frequências mais elevadas**, indicando melhor resposta cardiovascular  

📌 **Insight:** Frequências cardíacas máximas mais baixas podem estar associadas a limitações cardiovasculares, sendo um importante sinal de risco.


```python
plot_triplo(dados, 'Doenca cardiaca', 'Frequencia cardiaca max')
```

### 📉 Depressão do Segmento ST (`Depressao ST`)

A variável relacionada à depressão do segmento ST (indicador de isquemia) apresenta diferenças claras:

- Pacientes com doença cardíaca possuem **valores significativamente mais altos**  
- A distribuição mostra maior variabilidade e presença de valores extremos nesse grupo  

📌 **Insight:** A depressão do segmento ST é um dos indicadores mais fortes de anormalidade cardíaca, sendo altamente relevante para o diagnóstico.


```python
plot_triplo(dados, 'Doenca cardiaca', 'Depressao ST')
```

### 🩺 Número de Vasos Fluoroscopia (`Numero vasos fluro`)

A análise dessa variável evidencia um padrão bastante marcante:

- Pacientes com doença cardíaca apresentam **maior número de vasos afetados**  
- Já os indivíduos sem a doença tendem a concentrar-se em valores próximos de zero  

📌 **Insight:** O número de vasos afetados é um dos fatores mais discriminativos entre os grupos, indicando forte correlação com a presença da doença.


```python
plot_triplo(dados, 'Doenca cardiaca', 'Numero vasos fluro')
```

### 🧪 Teste de Cintilografia (`Teste cintilografia`)

Os resultados do teste de cintilografia também mostram separação relevante:

- Pacientes com doença cardíaca apresentam valores mais elevados e distribuídos em categorias específicas  
- Já os pacientes sem a doença tendem a concentrar-se em valores mais baixos  

📌 **Insight:** Esse exame demonstra alto poder de diferenciação, sendo um dos principais indicadores clínicos associados à doença cardíaca.


```python
plot_triplo(dados, 'Doenca cardiaca', 'Teste cintilografia')
```

## 🔗 Análise de Correlação com a Doença Cardíaca

```python
map_doenca = {
    'Presenca': 1,
    'Ausencia': 0
}
dados['Doenca cardiaca'] = dados['Doenca cardiaca'].map(map_doenca)
```

#### 🔴 Correlações Positivas (associadas à presença da doença)

- **Teste de cintilografia (~0.53)**  
- **Número de vasos fluoroscopia (~0.46)**  
- **Tipo de dor (~0.42)**  
- **Depressão ST (~0.42)**    

📌 **Interpretação:**  
Essas variáveis apresentam associação direta com a presença da doença cardíaca, indicando que valores mais altos tendem a aumentar a probabilidade de diagnóstico positivo.


```python
ax = plt.subplots(figsize=(15,15))
ax = sns.heatmap(dados.corr(), annot=True, cmap='coolwarm')
plt.title('Gráfico de correlação entre os Dados')
plt.tight_layout()
plt.show()
```
